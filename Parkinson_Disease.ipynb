```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from xgboost import XGBClassifier
import xgboost as xgb
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import joblib
```

```python
file = 'data.csv'
df = pd.read_csv(file)
df.head()
```

## Selecting The Features and Target Variables 

```python
X = df.drop(columns=['name', 'status'])
```

```python
y = df['status']
```

## EDA

```python
for data in X.columns:
    fig, ax = plt.subplots(figsize=(16,8))
    sns.boxplot(x=X[data])
    plt.title('Exploration Data Analysis')
plt.tight_layout()
plt.show()
```

```python
for col in X.columns:
    ax, fig = plt.subplots(figsize=(15,7))
    sns.scatterplot(x=X[col], y=y)
    plt.title('Exploration Data Analysis')
    plt.tight_layout()
    plt.show()
```

## With the Help of the EDA, we are able to detect that there are Outliers in our Features, So we have to Normalize Our Data so it can generate a better output

```python
sc = StandardScaler()
```

```python
X_scaled = sc.fit_transform(X)
```

## Splitting Our Data Into Training And Validation Set

```python
X_train, X_val, y_train, y_val = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
```

```python
X_train.shape, X_val.shape 
```

## Building Our Model

```python
model = XGBClassifier(n_estimators=60,
                     reg_lambda=0.06,
                     max_depth=6,
                     learning_rate=0.02)
```

```python
model.fit(X_train, y_train,
         eval_metric='error',
         verbose=True,
         early_stopping_rounds=10,
         eval_set=[(X_val, y_val)])
```

## Plotting The Importance Of The Features To The Model

```python
plt.figure(figsize=(15,10))
xgb.plot_importance(model, ax=plt.gca())
plt.show()
```

## Checking The Accuracy of the Model in the Testing set

```python
predict_X_test = model.predict(X_val)
```

```python
accuracy_score(predict_X_test, y_val)
```

## Testing Set Confusion matrix

```python
cf_matrix = confusion_matrix(predict_X_test, y_val)
```

```python
cf_matrix
```

## Testing Set Classification Report

```python
print(classification_report(predict_X_test, y_val))
```

```python
fig, ax = plt.subplots(1, figsize=(16,8))
sns.heatmap(cf_matrix, annot=True, fmt='', cmap='rainbow')

ax.set_title('Seaborn Confusion Matrix with labels\n\n');

## Display the visualization of the Confusion Matrix.
plt.tight_layout()
plt.show()
```

## Saving The Model Into a File

```python
joblib.dump(model, 'Parkison_Disease_Detection_Model.pkl')
```
